```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv('../datasets/datos05_train.csv')
X_orig = df.drop(columns="ActividadFisica")
y_orig = df[["ActividadFisica"]]

X_train, X_test, y_train, y_test = train_test_split(X_orig, y_orig, test_size=0.25, stratify=y_orig)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.neighbors import KNeighborsClassifier

# Definir el preprocesador
preprocesador = ColumnTransformer(
    transformers=[
        ("num", Pipeline(
            [
                ("imputador", SimpleImputer()),
                ("std", StandardScaler())
            ]
        ), ['Peso', 'Estatura', 'Edad', 'PctGrasaCorporal']),
        
        ("cat", Pipeline(
            [
                ("imputador", SimpleImputer(strategy="most_frequent")),
                ("one", OneHotEncoder(handle_unknown="ignore"))
            ]
        ), ["EstadoCivil"])
    ]
)

# Crear el modelo KNeighborsClassifier
modelo = KNeighborsClassifier(3)

# Crear el pipeline final
modelo_final = Pipeline([
    ("preprocesar", preprocesador),
    ("clasificar", modelo)
])
```

```python
modelo_final.fit(X_orig, y_orig)

modelo_final.predict(y_orig)
```

```python
from sklearn.metrics import accuracy_score

modelo_final.fit(X_train, y_train)

y_pred = modelo_final.predict(X_test)

accuracy_score(y_pred, y_test)
```

```python
from sklearn.model_selection import cross_val_score

accuracies = cross_val_score(modelo_final, X_orig, y_orig, cv=10)
```

```python
import numpy as np

print(np.mean(accuracies))
print(np.std(accuracies))
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_validate

cv = StratifiedKFold(n_splits=10, shuffle=True)

resultado = cross_validate(modelo_final, X_orig, y_orig, cv=cv, return_train_score=True)
resultado
```

```python
from sklearn.model_selection import GridSearchCV

df = pd.read_csv('../datasets/datos05_train.csv')
X_orig = df.drop(columns="ActividadFisica")
y_orig = df[["ActividadFisica"]]

X_train, X_test, y_train, y_test = train_test_split(X_orig, y_orig, test_size=0.25, stratify=y_orig)

# Definir el preprocesador
preprocesador = ColumnTransformer(
    transformers=[
        ("num", Pipeline(
            [
                ("imputador", SimpleImputer()),
                ("std", StandardScaler())
            ]
        ), ['Peso', 'Estatura', 'Edad', 'PctGrasaCorporal']),
        
        ("cat", Pipeline(
            [
                ("imputador", SimpleImputer(strategy="most_frequent")),
                ("one", OneHotEncoder(handle_unknown="ignore"))
            ]
        ), ["EstadoCivil"])
    ]
)

# Crear el modelo KNeighborsClassifier
modelo = KNeighborsClassifier(3)

# Crear el pipeline final
modelo_final = Pipeline([
    ("preprocesar", preprocesador),
    ("clasificar", modelo)
])
```

```python
# Definir un diccionario con los valores de hiperparámetros que queremos probar
param_grid = {"clasificar__n_neighbors": [1, 3, 5, 7, 9]}

# Crear objeto de CV
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)

# Crear un objeto de búsqueda
grid_search = GridSearchCV(modelo_final, param_grid, cv=cv, scoring="accuracy")
grid_search
```

```python
grid_search.fit(X_train, y_train)
```

```python
mejor_modelo = grid_search.best_estimator_

y_pred = mejor_modelo.predict(X_test)

accuracy_score(y_pred, y_test)

# Obtener el modelo con los mejores hiperparámetros. Ya está entrenado: GridSearchCV se ha encargado de ello porque usamos el valor por defecto refit=True
best_model = grid_search.best_estimator_

best_params = {key.replace("classifier__", ""): value for key, value in grid_search.best_params_.items()}
print("Mejores hiperparámetros:")
display(best_params)

# Predecir las etiquetas de los datos de test
y_pred = best_model.predict(X_test)

# Obtener el accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy final del modelo: {accuracy:.3f}")
```

# EJERCICIO FINAL: optimización de hiperparámetros y comparación de modelos

```python
df = pd.read_csv('../datasets/MathLearningDataset.csv')

X_orig = df.drop(columns=["Student ID", "Question ID", "Type of Answer"])
y_orig = df["Type of Answer"]

X_train, X_test, y_train, y_test = train_test_split(X_orig, y_orig, test_size=0.25, stratify=y_orig)
```

```python
# Definir el preprocesador
preprocesador = ColumnTransformer(
    transformers=[
        ("num", Pipeline(
            [
                ("imputador", SimpleImputer()),
                ("std", StandardScaler())
            ]
        ), ['Student Age', 'Response Time (s)']),
        
        ("cat", Pipeline(
            [
                ("imputador", SimpleImputer(strategy="most_frequent")),
                ("one", OneHotEncoder(handle_unknown="ignore"))
            ]
        ), ['Student Country', 'Question Level', 'Topic', 'Subtopic'])
    ]
)

# Crear el modelo KNeighborsClassifier
modelo = KNeighborsClassifier(3)

# Crear el pipeline final
modelo_final = Pipeline([
    ("preprocesar", preprocesador),
    ("clasificar", modelo)
])
```

```python
param_grid = {"clasificar__n_neighbors": [1, 3, 5, 7],
              "clasificar__p": [1, 2],
              "clasificar__weights": ["uniform", "distance"]}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)  # Crear objeto de CV

grid_search_knn = GridSearchCV(modelo_final, param_grid, cv=cv, scoring="accuracy")  # Crear un objeto de búsqueda

X_train, X_test, y_train, y_test = train_test_split(
    X_orig, y_orig, test_size=0.2, random_state=348, stratify=y_orig
)  # Dividir el dataset en entrenamiento y test

grid_search_knn.fit(X_train, y_train)  # Realizar la búsqueda sobre el conjunto de entrenamiento

# Mostrar los resultados para cada combinación de hiperparámetros
results = pd.DataFrame(grid_search_knn.cv_results_)

display(
    results[
        [
            "param_clasificar__n_neighbors",
            "param_clasificar__p",
            "param_clasificar__weights",
            "mean_test_score",
            "std_test_score",
            "rank_test_score",
        ]
    ].sort_values(by="rank_test_score")
)

print(f"Mejores hiperparámetros:")
grid_search_knn.best_params_
```

```python
param_grid = {
    "clasificar__max_depth": [1, 2, 3, None], # Cuando es None, no se restringe la profundidad (puede ser cualquiera)
    "clasificar__min_samples_split": [5, 10],  # Número mínimo de muestras requeridas para poder dividir un nodo interno
    "clasificar__min_samples_leaf": [5, 10,],  # Número mínimo de muestras requeridas para ser una hoja
    "clasificar__criterion": ["gini", "entropy"],  # Criterio para la división
    "clasificar__max_leaf_nodes": [3, 5, 10, None]  # Restringir el número máximo de nodos hoja
}

# Crear el pipeline final
modelo_final = Pipeline([
    ("preprocesar", preprocesador),
    ("clasificar", DecisionTreeClassifier(random_state=0))
])

# Número de combinaciones posibles
combinaciones = 1
for key, value in param_grid.items():
    combinaciones *= len(value)
print(f"Número de combinaciones posibles: {combinaciones}")

# Configurar CV y grid search
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=0)
grid_search_tree = GridSearchCV(modelo_final, param_grid, cv=cv, scoring="accuracy")  

grid_search_tree.fit(X_train, y_train) # Sobre los datos sin escalar porque ahora esto lo realiza el pipeline

# Mostrar los resultados para cada combinación de hiperparámetros

keys_to_keep = ["mean_test_score", "std_test_score", "rank_test_score"]
keys_to_keep += [key for key in grid_search_tree.cv_results_ if key.startswith("param_")]
filtered_results = {key: grid_search_tree.cv_results_[key] for key in keys_to_keep}
cleaned_results = {
    key.replace("param_clasificar__", "") if key.startswith("param_clasificar__") else key: value
    for key, value in filtered_results.items()
}

results = pd.DataFrame(cleaned_results)

display(results.sort_values(by="rank_test_score").head(10))
```

```python
# Obtener el modelo con los mejores hiperparámetros. Ya está entrenado: GridSearchCV se ha encargado de ello porque usamos el valor por defecto refit=True
best_model_knn = grid_search_knn.best_estimator_

best_params = {key.replace("classifier__", ""): value for key, value in grid_search_knn.best_params_.items()}
print("Mejores hiperparámetros:")
display(best_params)

# Predecir las etiquetas de los datos de test
y_pred = best_model.predict(X_test)

# Obtener el accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy final del modelo: {accuracy:.3f}")
```
